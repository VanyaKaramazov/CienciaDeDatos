```python
import pandas as pd
from scipy import stats

# --- 1. LIMPIEZA DE DATOS ---
# Importamos la base de datos que descargué de Kaggle
data = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')

print("Valores nulos por columna antes de limpieza:\n", data.isnull().sum())

# Eliminamos datos nulos como pide la instrucción
data_2 = data.dropna()


# --- 2. PRUEBA DE HIPÓTESIS ---
# Pregunta: ¿La duración del sueño es diferente entre hombres y mujeres?
# H0 (Hipótesis Nula): No hay diferencia en las medias.
# H1 (Hipótesis Alternativa): Sí hay diferencia.

# Filtro las duraciones de sueño por género
hombres = data_2[data_2['Gender'] == 'Male']['Sleep Duration']
mujeres = data_2[data_2['Gender'] == 'Female']['Sleep Duration']

# T-test de muestras independientes
t_stat, p_value = stats.ttest_ind(hombres, mujeres)

print("\n--- Prueba de Hipótesis (T-test) ---")
print(f"Estadístico t: {t_stat:.4f}")
print(f"Valor p (p-value): {p_value:.4e}")


# --- 3. ANÁLISIS DE CORRELACIÓN ---
correlacion = data_2['Sleep Duration'].corr(data_2['Stress Level'])
print("\n--- Correlación ---")
print(f"Coeficiente de Pearson: {correlacion:.4f}")


# --- 4. ANÁLISIS DE COVARIANZA ---
covarianza = data_2['Sleep Duration'].cov(data_2['Stress Level'])
print("\n--- Covarianza ---")
print(f"Covarianza: {covarianza:.4f}")
```

Reflexión de los análisis realizados:

Prueba de Hipótesis: Al aplicar el T-test para comparar la duración del sueño entre hombres y mujeres, evalué el p-value obtenido contra un nivel de significancia estándar (0.05). Esto me permite concluir estadísticamente si existe una diferencia real en las horas de descanso entre ambos géneros en la muestra analizada, o si cualquier variación se debe simplemente al azar.

Correlación y Covarianza: El coeficiente de correlación arrojó un valor negativo, lo que indica una relación inversa entre ambas variables. Esto sugiere fuertemente que, a medida que los individuos duermen menos horas, sus niveles de estrés reportados tienden a ser más altos. El valor negativo de la covarianza respalda matemáticamente esta conclusión, indicando que el sueño y el estrés se mueven en direcciones opuestas.

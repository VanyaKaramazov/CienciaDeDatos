```python
# Importamos las librerías que usaremos en esta sesión
import matplotlib.pyplot as plt
import pandas as pd
```

## 1. ¿Por qué usar Pandas?

En la vida real, los datos no siempre vienen en listas separadas.

Muchas veces trabajamos con información en forma de tabla:
- nombres de productos
- ventas
- edades
- calificaciones
- presupuestos

Pandas nos ayuda a organizar esa información en una estructura llamada `DataFrame`.


```python
# Creamos un diccionario con datos de ejemplo
datos = {
    'Producto': ['Cuaderno', 'Lápiz', 'Borrador', 'Regla'],
    'Ventas': [30, 50, 20, 15]
}

# Convertimos el diccionario en un DataFrame
df = pd.DataFrame(datos)

# Mostramos la tabla
df
```

## 2. Gráfico de barras

El gráfico de barras es útil cuando queremos comparar cantidades entre categorías.

Por ejemplo:
- ventas por producto
- alumnos por grupo
- presupuesto por área
- población por estado


```python
# Creamos una figura y un eje
fig, ax = plt.subplots()

# Usamos una columna para el eje X y otra para el eje Y
ax.bar(df['Producto'], df['Ventas'])

# Agregamos título y etiquetas
ax.set_title('Ventas por producto')
ax.set_xlabel('Producto')
ax.set_ylabel('Cantidad vendida')

# Mostramos la gráfica
plt.show()
```

## 3. Personalizar un gráfico de barras

Podemos mejorar la apariencia de la gráfica cambiando colores, bordes y tamaño de figura.


```python
# Creamos una figura más amplia
fig, ax = plt.subplots(figsize=(8,4))

# Dibujamos las barras con color y borde
ax.bar(df['Producto'], df['Ventas'], color= "#8786BD" , edgecolor='black')

# Agregamos título y etiquetas
ax.set_title('Ventas por producto')
ax.set_xlabel('Producto')
ax.set_ylabel('Cantidad vendida')

# Mostramos la gráfica
plt.show()
```

## 4. Gráfico de dispersión

El gráfico de dispersión sirve para observar la relación entre dos variables numéricas.

Por ejemplo:
- horas de estudio y calificación
- edad y salario
- temperatura y ventas


```python
# Creamos un nuevo conjunto de datos
datos_estudiantes = {
    'Horas_estudio': [1, 2, 3, 4, 5, 6, 7],
    'Calificacion': [55, 60, 65, 70, 78, 85, 90]
}

# Convertimos los datos en DataFrame
df_estudiantes = pd.DataFrame(datos_estudiantes)

# Mostramos la tabla
df_estudiantes
```

```python
# Creamos la figura y el eje
fig, ax = plt.subplots()

# Dibujamos el gráfico de dispersión
ax.scatter(df_estudiantes['Horas_estudio'], df_estudiantes['Calificacion'])

# Agregamos título y etiquetas
ax.set_title('Horas de estudio y calificación')
ax.set_xlabel('Horas de estudio')
ax.set_ylabel('Calificación')

# Mostramos la gráfica
plt.show()
```

## 5. Personalizar un gráfico de dispersión

También podemos cambiar el color y el tamaño de los puntos para que la gráfica sea más clara.


```python
# Creamos la figura y el eje
fig, ax = plt.subplots(figsize=(8, 8))

# Dibujamos la dispersión con color y tamaño personalizados
ax.scatter(
    df_estudiantes['Horas_estudio'],
    df_estudiantes['Calificacion'],
    color='purple',
    s=200
)

# Agregamos título y etiquetas
ax.set_title('Relación entre estudio y calificación')
ax.set_xlabel('Horas de estudio')
ax.set_ylabel('Calificación')

# Mostramos la gráfica
plt.show()
```

## 6. Comparación entre barras y dispersión

Usamos barras cuando queremos comparar categorías.

Usamos dispersión cuando queremos analizar la relación entre dos variables numéricas.

### Ejemplo:
- Barras: ventas de 4 productos.
- Dispersión: horas de estudio y calificación.

```python
#Solución al Ejercicio 1
datos = {"materia": ["Matemáticas", "Historia", "Biología", "Inglés"],
         "promedio": [8.5, 7.8, 9.1, 8.0]
         }
#Creamos el dataframe
df = pd.DataFrame(datos)

# Creamos la figura
fig, ax = plt.subplots()

# Dibujamos las barras con color y borde
ax.bar(df['materia'], df['promedio'], color= "#8786BD" , edgecolor='black')

# Agregamos título y etiquetas
ax.set_title('Promedio por materias')
ax.set_xlabel('Materia')
ax.set_ylabel('Promedio')

# Mostramos la gráfica
plt.show()

```

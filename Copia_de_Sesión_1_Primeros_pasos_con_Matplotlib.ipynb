````python
# Importamos matplotlib.pyplot
# Usamos el alias 'plt' porque es la convención más común
import matplotlib.pyplot as plt
````

## 1. ¿Qué es Matplotlib?

Matplotlib es una librería de Python que nos permite crear gráficas para visualizar datos.

En esta primera clase trabajaremos con una gráfica de líneas, que sirve para mostrar cambios, tendencias o comparaciones sencillas.

Por ejemplo:
- ventas por día
- temperatura por semana
- asistencia por mes


## 2. Nuestra primera gráfica

Primero vamos a trabajar con dos listas:

- Una lista para el eje horizontal (X)
- Una lista para el eje vertical (Y)

En este ejemplo, mostraremos las ventas de una tienda durante cinco días.


````python
# Creamos los datos que vamos a graficar
dias = ['Lun', 'Mar', 'Mié', 'Jue', 'Vie']
ventas = [10, 15, 8, 20, 18]

# Dibujamos una gráfica de líneas
plt.plot(dias, ventas)

# Mostramos la gráfica en pantalla
plt.show()
````

## 3. Agregar contexto a la gráfica

Una gráfica necesita contexto para que sea fácil de entender.

Por eso, normalmente agregamos:
- un título
- una etiqueta para el eje X
- una etiqueta para el eje Y


````python
# Creamos nuevamente los datos
dias = ['Lun', 'Mar', 'Mié', 'Jue', 'Vie']
ventas = [10, 15, 8, 20, 18]

# Dibujamos la gráfica
plt.plot(dias, ventas)

# Agregamos un título
plt.title('Ventas por día')

# Agregamos etiquetas a los ejes
plt.xlabel('Días de la semana')
plt.ylabel('Cantidad de ventas')

# Mostramos la gráfica final
plt.show()
````

## 4. La forma recomendada: `fig` y `ax`

Matplotlib tiene una forma más organizada de trabajar.

Cuando escribimos:

```python
fig, ax = plt.subplots()
```

estamos creando:

- `fig`: la figura completa
- `ax`: el área donde se dibuja la gráfica

Este enfoque es muy útil porque nos da más control y será el más usado en el curso.


````python
# Datos del ejemplo
dias = ['Lun', 'Mar', 'Mié', 'Jue', 'Vie']
ventas = [10, 15, 8, 20, 18]

# Creamos la figura y los ejes
fig, ax = plt.subplots()

# Dibujamos la gráfica en el objeto ax
ax.plot(dias, ventas)

# Personalizamos la gráfica
ax.set_title('Ventas por día')
ax.set_xlabel('Días de la semana')
ax.set_ylabel('Cantidad de ventas')

# Mostramos el resultado
plt.show()
````

## 5. Comparación rápida entre los dos estilos

Hasta ahora vimos dos formas de crear una gráfica:

### Estilo rápido
Usa funciones como:
- `plt.plot()`
- `plt.title()`
- `plt.xlabel()`
- `plt.ylabel()`

### Estilo recomendado
Usa objetos:
- `fig, ax = plt.subplots()`
- `ax.plot()`
- `ax.set_title()`
- `ax.set_xlabel()`
- `ax.set_ylabel()`

En este curso usaremos ambos, pero daremos preferencia al segundo.


## 6. Ejemplo completo comentado

Ahora veremos un ejemplo completo con comentarios paso a paso.


````python
# Importamos matplotlib
import matplotlib.pyplot as plt

# Lista con los meses
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May']

# Lista con la cantidad de usuarios registrados
usuarios = [100, 120, 140, 180, 210]

# Creamos la figura y el área de dibujo
fig, ax = plt.subplots()

# Dibujamos la gráfica de líneas
# marker='o' agrega un círculo en cada punto de datos
ax.plot(meses, usuarios, marker='D')

# Agregamos un título a la gráfica
ax.set_title('Crecimiento de usuarios')

# Etiquetamos el eje X
ax.set_xlabel('Meses')

# Etiquetamos el eje Y
ax.set_ylabel('Número de usuarios')

# Mostramos la gráfica
plt.show()
````

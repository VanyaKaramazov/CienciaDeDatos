## 🧹 Sección 3 — Los 5 Poderes del Texto: Limpieza en Python Puro
*(Sesiones 6 y 13)*

El sistema de captura de datos entregó registros sucios.
Limpia cada uno usando los poderes del texto.

| Método | Qué hace | Ejemplo |
|---|---|---|
| `.lower()` | Todo a minúsculas | `"GRIPE"` → `"gripe"` |
| `.upper()` | Todo a mayúsculas | `"norte"` → `"NORTE"` |
| `.title()` | Primera letra de cada palabra en mayúscula | `"juan prez"` → `"Juan Prez"` |
| `.strip()` | Quita espacios al inicio y al final | `"  covid  "` → `"covid"` |
| `.replace(v, n)` | Sustituye texto | `"Tipo 2"` → `"Tipo II"` |


```python
# ─────────────────────────────────────────────────────
# EJERCICIO 3.1 — Limpia cada campo individualmente
# Regla de oro: siempre SOBRESCRIBE la variable para guardar el cambio
# ─────────────────────────────────────────────────────
nombre_sucio  = "  JUAN PABLO REYES "    # espacios + todo en mayúsculas
plantel_sucio = "plantel SUR "           # mezcla de casos + espacio al final
materia_sucia = "MATEMATICAS tipo 2"     # mayúsculas + necesita estandarizar
correo_sucio  = "  AlUmNo@EDUDATA.com "  # caos total

# Limpieza
nombre_limpio  = nombre_sucio.strip().title()          # strip + title
plantel_limpio = plantel_sucio.strip().title()          # strip + title
materia_limpia = materia_sucia.strip().replace("tipo 2", "Tipo II")  # strip + replace
correo_limpio  = correo_sucio.strip().lower()           # strip + lower

print("=== Datos Limpios ===")
print(f"  Nombre   : {nombre_limpio}")
print(f"  Plantel  : {plantel_limpio}")
print(f"  Materia  : {materia_limpia}")
print(f"  Correo   : {correo_limpio}")

# ─────────────────────────────────────────────────────
# EJERCICIO 3.2 — Encadenamiento en una línea (estilo profesional)
# ─────────────────────────────────────────────────────
nombre_pro = "  SOFÍA ramírez ".strip().title()
print(f"\nEncadenado en una línea: {nombre_pro}")

# ─────────────────────────────────────────────────────
# EJERCICIO 3.3 — Automatiza la limpieza de una lista completa con for
# Aplica .strip().lower() a cada ciudad para estandarizar
# ─────────────────────────────────────────────────────
planteles_sucios = ["Plantel Norte ", "plantel sur", "PLANTEL NORTE", "Plantel Sur  "]
planteles_limpios = []

for plantel in planteles_sucios:
    limpio = plantel.strip().lower()    # strip + lower
    planteles_limpios.append(limpio)

print(f"\nPlanteles sucios : {planteles_sucios}")
print(f"Planteles limpios: {planteles_limpios}")

# ¿Cuántos alumnos son del plantel norte?
print(f"Alumnos plantel norte: {planteles_limpios.count('plantel norte')}")

# ─────────────────────────────────────────────────────
# EJERCICIO 3.4 — Limpia un diccionario (múltiples campos a la vez)
# ─────────────────────────────────────────────────────
registro_sucio = {
    "nombre"   : "  CARLOS VEGA ",
    "plantel"  : "plantel norte ",
    "materia"  : "ALGEBRA  ",
}

registro_sucio["nombre"]  = registro_sucio["nombre"].strip().title()   # strip + title
registro_sucio["plantel"] = registro_sucio["plantel"].strip().lower()  # strip + lower
registro_sucio["materia"] = registro_sucio["materia"].strip().title()  # strip + title

print(f"\nRegistro limpio: {registro_sucio}")
```

## 🔄 Sección 7 — Bucles for / while y Funciones
*(Sesiones 7 y 8)*


```python
# Base de datos de alumnos del mes
alumnos_mes = [
    {"nombre":"Valeria Mendoza",  "matricula":"EDU-047","promedio":6.8, "faltas":5, "horas_estudio":3.0,"aprueba":False,"plantel":"Norte"},
    {"nombre":"Andrés Jiménez",   "matricula":"EDU-023","promedio":4.5, "faltas":9, "horas_estudio":1.5,"aprueba":False,"plantel":"Sur"},
    {"nombre":"Sofía Ramírez",    "matricula":"EDU-088","promedio":9.2, "faltas":1, "horas_estudio":6.0,"aprueba":True, "plantel":"Norte"},
    {"nombre":"Luis Torres",      "matricula":"EDU-031","promedio":7.8, "faltas":3, "horas_estudio":4.5,"aprueba":True, "plantel":"Centro"},
    {"nombre":"Camila Vega",      "matricula":"EDU-055","promedio":5.1, "faltas":8, "horas_estudio":2.0,"aprueba":False,"plantel":"Centro"},
    {"nombre":"Roberto Cruz",     "matricula":"EDU-072","promedio":8.6, "faltas":2, "horas_estudio":5.5,"aprueba":True, "plantel":"Sur"},
    {"nombre":"Andrea Morales",   "matricula":"EDU-019","promedio":3.8, "faltas":11,"horas_estudio":0.5,"aprueba":False,"plantel":"Norte"},
    {"nombre":"Martín Soto",      "matricula":"EDU-064","promedio":7.5, "faltas":4, "horas_estudio":4.0,"aprueba":True, "plantel":"Sur"},
]

# ─────────────────────────────────────────────────────
# EJERCICIO 7.1 — Función reutilizable de clasificación
# ─────────────────────────────────────────────────────
def clasificar_riesgo(promedio, faltas):
    """Devuelve el nivel de riesgo académico."""
    if promedio >= 8.5:
        nivel = "🟢 EXCELENTE"
    elif promedio >= 7.0:
        nivel = "🟡 REGULAR"
    else:
        nivel = "🔴 EN RIESGO"

    if faltas > 7 and nivel != "🔴 EN RIESGO":
        nivel = "🔴 EN RIESGO"
    return nivel

# Prueba
print(clasificar_riesgo(9.2, 1))    # 🟢 EXCELENTE
print(clasificar_riesgo(7.8, 8))    # 🔴 EN RIESGO (penalización por faltas)
print(clasificar_riesgo(3.8, 11))   # 🔴 EN RIESGO

# ─────────────────────────────────────────────────────
# EJERCICIO 7.2 — Bucle for: reporte académico completo
# ─────────────────────────────────────────────────────
print("\n=== REPORTE ACADÉMICO — EDUDATA ===")
print(f"{'#':<4}{'Alumno':<22}{'Prom':>6}{'Faltas':>7}  Nivel")
print("-" * 58)

en_riesgo = 0
for i, a in enumerate(alumnos_mes, 1):
    nivel = clasificar_riesgo(a["promedio"], a["faltas"])
    print(f"{i:<4}{a['nombre']:<22}{a['promedio']:>5.1f} {a['faltas']:>5}f  {nivel}")
    if "RIESGO" in nivel:
        en_riesgo += 1

print("-" * 58)
print(f"Total alumnos      : {len(alumnos_mes)}")
print(f"Alumnos en riesgo  : {en_riesgo} ({en_riesgo/len(alumnos_mes)*100:.1f}%)")

# ─────────────────────────────────────────────────────
# EJERCICIO 7.3 — while: acumula horas de estudio hasta llegar a la meta
# ─────────────────────────────────────────────────────
print("\n=== Alumnos con ≥4h estudio diario (meta de rendimiento) ===")
meta_horas = 4.0
acumulado  = 0
i = 0

while i < len(alumnos_mes):
    if alumnos_mes[i]["horas_estudio"] >= meta_horas:
        acumulado += 1
        print(f"  ✅ {alumnos_mes[i]['nombre']} — {alumnos_mes[i]['horas_estudio']}h/día")
    i += 1

print(f"Total cumplen la meta: {acumulado} de {len(alumnos_mes)} alumnos")
```

## 📐 Sección 8 — Estadística Manual: Promedio, Mediana, Rango y Moda
*(Sesión 9)*

Calcula las estadísticas **desde cero con Python puro** antes de usar librerías.


```python
# ─────────────────────────────────────────────────────
# EJERCICIO 8.1 — Promedio manual (sum / len + round)
# ─────────────────────────────────────────────────────
promedios = [a["promedio"] for a in alumnos_mes]

suma       = sum(promedios)
cantidad   = len(promedios)
prom_manual= round(suma / cantidad, 1)   # redondea a 1 decimal

print(f"Promedios del grupo   : {promedios}")
print(f"Suma total            : {suma}")
print(f"Número de alumnos     : {cantidad}")
print(f"Promedio manual       : {prom_manual}")

# ─────────────────────────────────────────────────────
# EJERCICIO 8.2 — Mediana manual (sorted + índice central)
# ─────────────────────────────────────────────────────
promedios_ord = sorted(promedios)
idx_centro    = len(promedios_ord) // 2
mediana_manual= promedios_ord[idx_centro]

print(f"\nPromedios ordenados   : {promedios_ord}")
print(f"Índice central        : {idx_centro}")
print(f"Mediana               : {mediana_manual}")

# ─────────────────────────────────────────────────────
# EJERCICIO 8.3 — Rango y detección de outlier
# ─────────────────────────────────────────────────────
maximo = max(promedios)
minimo = min(promedios)
rango  = round(maximo - minimo, 1)

print(f"\nPromedio máximo       : {maximo}")
print(f"Promedio mínimo       : {minimo}")
print(f"Rango                 : {rango}")

# ─────────────────────────────────────────────────────
# EJERCICIO 8.4 — Moda manual (diccionario contador)
# ─────────────────────────────────────────────────────
faltas_lista = [a["faltas"] for a in alumnos_mes]
conteo = {}

for valor in faltas_lista:
    if valor in conteo:
        conteo[valor] += 1
    else:
        conteo[valor] = 1   # primera vez que aparece

moda_faltas = max(conteo, key=conteo.get)
print(f"\nTabla de frecuencias faltas: {conteo}")
print(f"Moda (faltas más frecuente): {moda_faltas} falta(s) — aparece {conteo[moda_faltas]} vez/veces")
```

## 🧮 Sección 9 — Módulo `statistics`: Automatizando el Análisis
*(Sesión 10)*


```python
import statistics

promedios     = [a["promedio"]       for a in alumnos_mes]
faltas        = [a["faltas"]         for a in alumnos_mes]
horas_estudio = [a["horas_estudio"]  for a in alumnos_mes]

# ─────────────────────────────────────────────────────
# EJERCICIO 9.1 — mean, median, mode
# ─────────────────────────────────────────────────────
prom_auto  = statistics.mean(promedios)
med_auto   = statistics.median(promedios)
moda_auto  = statistics.mode(faltas)

print("=== Análisis Estadístico — EduData ===")
print(f"  Promedio grupal (mean)   : {round(prom_auto, 2)}")
print(f"  Mediana promedios        : {med_auto}")
print(f"  Moda faltas              : {moda_auto} falta(s) al mes")

# ─────────────────────────────────────────────────────
# EJERCICIO 9.2 — stdev y % de variabilidad
# ─────────────────────────────────────────────────────
desv_prom  = statistics.stdev(promedios)
desv_horas = statistics.stdev(horas_estudio)
pct_variab = round((desv_prom / prom_auto) * 100, 1)

print(f"\n  Desv. estándar promedios : {round(desv_prom, 2)}")
print(f"  % variabilidad           : {pct_variab}%")

if pct_variab > 20:
    print("  ⚠️  Alta variabilidad — el grupo es muy heterogéneo.")
else:
    print("  ✅ Baja variabilidad — el grupo es consistente.")

# Comparación promedio vs mediana
dif = abs(prom_auto - med_auto)
print(f"\n  Diferencia prom-mediana  : {round(dif, 2)}")
if dif > 1.0:
    print("  ⚠️  Posible sesgo. Usa la MEDIANA para reportar al director.")

# ─────────────────────────────────────────────────────
# EJERCICIO 9.3 — multimode: ¿hay empate?
# ─────────────────────────────────────────────────────
todas_modas = statistics.multimode(faltas)
print(f"\n  Todas las modas de faltas: {todas_modas}")
```

## 🔍 Sección 10 — Los 4 Sospechosos + Detección Programática
*(Sesiones 11 y 12)*

Aplica el principio **GIGO** (*Garbage In, Garbage Out*) antes de cualquier análisis.


```python
# Dataset SUCIO exportado del sistema de control escolar
registros_sucios = [
    {"id":1,  "nombre":"Ana López",    "promedio":8.5, "plantel":"Norte"},
    {"id":2,  "nombre":"pedro ramos",  "promedio":6.2, "plantel":"sur"},        # inconsistencia
    {"id":3,  "nombre":"Carlos Vera",  "promedio":None,"plantel":"Centro"},     # valor nulo
    {"id":4,  "nombre":"Luisa Pérez",  "promedio":95.0,"plantel":"Norte"},      # outlier (escala 0-10)
    {"id":5,  "nombre":"Rosa Vega",    "promedio":7.4, "plantel":"Sur"},
    {"id":2,  "nombre":"pedro ramos",  "promedio":6.2, "plantel":"sur"},        # duplicado
]

# ─────────────────────────────────────────────────────
# EJERCICIO 10.1 — ¿Qué pasa si ignoras los datos sucios?
# ─────────────────────────────────────────────────────
import statistics

try:
    proms_sucios = [r["promedio"] for r in registros_sucios]
    prom = statistics.mean(proms_sucios)
    print(f"Promedio (sucio): {prom}")
except TypeError as e:
    print(f"❌ ERROR: {e}")
    print("   None rompe statistics.mean — hay que limpiar ANTES de analizar.")

# ─────────────────────────────────────────────────────
# EJERCICIO 10.2 — Radar de nulos
# ─────────────────────────────────────────────────────
print("\n=== Radar de Nulos ===")
if None in [r["promedio"] for r in registros_sucios]:
    print("  🚨 ALERTA: Se detectaron valores None en la columna 'promedio'")

total_nulos = 0
for pos, r in enumerate(registros_sucios):
    if r["promedio"] is None:    # is None
        print(f"  ⚠️  Nulo en posición {pos} — Alumno: {r['nombre']}")
        total_nulos += 1
print(f"  Total nulos: {total_nulos}")

# ─────────────────────────────────────────────────────
# EJERCICIO 10.3 — Radar de duplicados (lista de control + set)
# ─────────────────────────────────────────────────────
print("\n=== Radar de Duplicados ===")
ids_vistos, duplicados_cnt = [], 0

for r in registros_sucios:
    if r["id"] in ids_vistos:
        print(f"  🔁 Duplicado: ID={r['id']} ({r['nombre']})")
        duplicados_cnt += 1
    else:
        ids_vistos.append(r["id"])

ids_unicos = list(set([r["id"] for r in registros_sucios]))
print(f"  Duplicados: {duplicados_cnt} | IDs únicos: {sorted(ids_unicos)}")

# ─────────────────────────────────────────────────────
# EJERCICIO 10.4 — Identifica los 4 sospechosos
# ─────────────────────────────────────────────────────
nulos, outliers, inconsistencias = [], [], []
ids = [r["id"] for r in registros_sucios]
duplicados = [i for i in set(ids) if ids.count(i) > 1]

for r in registros_sucios:
    if r["promedio"] is None:
        nulos.append(r["id"])
    elif isinstance(r["promedio"], float) and (r["promedio"] < 0 or r["promedio"] > 10):
        outliers.append(r["id"])
    if r["nombre"][0].islower() or r["plantel"][0].islower():
        inconsistencias.append(r["id"])

print("\n=== Reporte de los 4 Sospechosos ===")
print(f"  🔴 Nulos          : IDs {nulos}")
print(f"  🔁 Duplicados     : IDs {duplicados}")
print(f"  📊 Outliers       : IDs {outliers}")
print(f"  ✏️  Inconsistencias: IDs {inconsistencias}")
```

## 🐼 Sección 11 — Limpieza Profesional con Pandas
*(Sesiones 13 y 14)*

Ahora aplica los 6 comandos de inspección y el pipeline de limpieza.

| Comando | Qué te dice |
|---|---|
| `df.head(n)` | Primeras n filas |
| `df.tail(n)` | Últimas n filas |
| `df.shape` | Dimensiones (filas, columnas) |
| `df.info()` | Tipos de datos + conteo de no-nulos |
| `df.describe()` | Estadísticas de columnas numéricas |
| `df.isnull().sum()` | Conteo exacto de nulos por columna |


```python
import pandas as pd

# Dataset del instituto — con errores intencionales
datos_instituto = {
    "ID"           : [101, 102, 103, 104, 102, 105, 106],
    "Nombre"       : ["Ana García","Luis Ramos","Carmen Vega","Jorge Díaz",
                      "Luis Ramos","Sofía Luna","Pedro Cruz"],
    "Promedio"     : [8.5, 6.2, None, 7.8, 6.2, None, 9.1],
    "Faltas"       : [2, 5, 8, 3, 5, 7, 1],
    "Plantel"      : ["norte","NORTE","Sur ","  centro  ","NORTE","sur","Centro"],
}

df = pd.DataFrame(datos_instituto)

# ─────────────────────────────────────────────────────
# EJERCICIO 11.1 — Los 6 comandos de inspección
# ─────────────────────────────────────────────────────
print("=== 1. Primeras filas ===")
print(df.head())

print("\n=== 2. Últimas 3 filas ===")
print(df.tail(3))

print("\n=== 3. Dimensiones ===")
print(f"  {df.shape[0]} filas × {df.shape[1]} columnas")

print("\n=== 4. Info (tipos + no-nulos) ===")
df.info()

print("\n=== 5. Estadísticas numéricas ===")
print(df.describe().round(2))

print("\n=== 6. Escáner de nulos ===")
print(df.isnull().sum())

# ─────────────────────────────────────────────────────
# EJERCICIO 11.2 — Identifica duplicados antes de borrar
# ─────────────────────────────────────────────────────
print("\n=== Identificación de Duplicados ===")
print(f"  Total duplicados: {df.duplicated().sum()}")
print(df[df.duplicated(keep=False)])   # muestra TODAS las copias

# ─────────────────────────────────────────────────────
# EJERCICIO 11.3 — Pipeline de limpieza en orden
# ─────────────────────────────────────────────────────
# PASO 1: Elimina duplicados (guarda el primero, borra la copia)
df = df.drop_duplicates()
print(f"\nDespués de dropduplicates: {df.shape[0]} filas")

# PASO 2: Imputa nulos de Promedio con la MEDIANA (mejor que el promedio ante outliers)
mediana_prom = df["Promedio"].median()
df["Promedio"] = df["Promedio"].fillna(mediana_prom)
print(f"Mediana usada para imputar: {mediana_prom}")

# PASO 3: Estandariza la columna Plantel con .str.strip().str.lower()
df["Plantel"] = df["Plantel"].str.strip().str.lower()
print(f"Planteles estandarizados: {df['Plantel'].unique()}")

# PASO 4: Verifica que no queden nulos
print(f"\nNulos restantes:\n{df.isnull().sum()}")
print("\n=== Dataset Limpio Final ===")
print(df)
```

## 🏆 Reto Final — Pipeline Integrador Completo S1–S16
*(Integrador de todas las sesiones)*

Completa la función `pipeline_edudata()` que ejecuta las **5 fases** del ciclo completo
de ciencia de datos aprendido en las sesiones 1 a 16.


```python
def pipeline_edudata(lista_alumnos):
    """
    Pipeline completo de Ciencia de Datos — Instituto EduData.
    Integra sesiones 1-16 del Diplomado.
    """
    import statistics, numpy as np
    import pandas as pd

    print("\n" + "="*62)
    print("  🔤  FASE 1 — LIMPIEZA DE TEXTO (S6 y S13)")
    print("="*62)
    # Aplica .strip().title() al nombre de cada alumno
    for a in lista_alumnos:
        a["nombre"] = a["nombre"].strip().title()   # strip + title
    print(f"  Nombres estandarizados: {[a['nombre'] for a in lista_alumnos]}")

    print("\n" + "="*62)
    print("  🛡️   FASE 2 — AUDITORÍA DE DATOS (S11 y S12)")
    print("="*62)
    ids_vistos, nulos, duplicados_cnt = [], 0, 0
    for a in lista_alumnos:
        if a["promedio"] is None:
            print(f"  🔴 Nulo: {a['nombre']} — promedio faltante")
            nulos += 1
        if a["matricula"] in ids_vistos:
            print(f"  🔁 Duplicado: {a['matricula']} ({a['nombre']})")
            duplicados_cnt += 1
        else:
            ids_vistos.append(a["matricula"])
    print(f"  Nulos: {nulos} | Duplicados: {duplicados_cnt}")

    print("\n" + "="*62)
    print("  📐  FASE 3 — ESTADÍSTICA (S9 y S10)")
    print("="*62)
    validos    = [a for a in lista_alumnos if a["promedio"] is not None]
    proms      = [a["promedio"] for a in validos]
    hrs        = [a["horas_estudio"] for a in validos]

    # Estadística manual (S9)
    prom_manual  = round(sum(proms) / len(proms), 2)
    sorted_p     = sorted(proms)
    mediana_man  = sorted_p[len(sorted_p) // 2]

    # Módulo statistics (S10)
    prom_auto    = statistics.mean(proms)
    desv_auto    = statistics.stdev(hrs)

    print(f"  Promedio manual   : {prom_manual}")
    print(f"  Promedio auto     : {round(prom_auto, 2)}")
    print(f"  Mediana promedios : {mediana_man}")
    print(f"  Desv. estándar hrs: {round(desv_auto, 2)}")

    print("\n" + "="*62)
    print("  🚦  FASE 4 — CLASIFICACIÓN (S6, S7 y S8)")
    print("="*62)
    en_riesgo, excelentes = 0, 0
    for a in validos:
        nivel = clasificar_riesgo(a["promedio"], a["faltas"])
        print(f"  {a['nombre']:<22} → {nivel}")
        if "RIESGO"    in nivel: en_riesgo  += 1
        if "EXCELENTE" in nivel: excelentes += 1

    print("\n" + "="*62)
    print("  📊  FASE 5 — PANDAS (S14): Resumen por plantel")
    print("="*62)
    df_temp = pd.DataFrame(validos)
    df_temp["nivel"] = df_temp.apply(
        lambda row: clasificar_riesgo(row["promedio"], row["faltas"]), axis=1)
    print(df_temp.groupby("plantel")["promedio"].mean().round(2))

    print("\n" + "="*62)
    print("  📋  RESUMEN EJECUTIVO (f-strings)")
    print("="*62)
    pct_riesgo = en_riesgo / len(validos) * 100
    print(f"  Total alumnos    : {len(lista_alumnos)}")
    print(f"  Datos válidos    : {len(validos)}")
    print(f"  Promedio grupal  : {round(prom_auto, 2)}/10")
    print(f"  Mediana          : {mediana_man}/10")
    print(f"  Alumnos excelentes: {excelentes}")
    print(f"  Alumnos en riesgo : {en_riesgo} ({pct_riesgo:.1f}%)")
    print("="*62)

# ── Ejecuta el pipeline ────────────────────────────────────────────
pipeline_edudata(alumnos_mes)
```
